```python
import random

# =========================
# ALIEN LANGUAGE GENERATOR
# =========================
syllables = ["ka", "zu", "ra", "xi", "lo", "tar", "un", "vek", "shi", "mor"]

def alien_word(min_syl=2, max_syl=4):
    return "".join(random.choice(syllables) for _ in range(random.randint(min_syl, max_syl))).capitalize()

# =========================
# PLANET GENERATOR
# =========================
planet_types = [
    "Desert World",
    "Ocean World",
    "Ice World",
    "Lava World",
    "Forest World",
    "Gas Giant",
    "Crystal World"
]

def generate_planet():
    return {
        "name": alien_word(),
        "type": random.choice(planet_types),
        "moons": random.randint(0, 4)
    }

# =========================
# ALIEN SPECIES GENERATOR
# =========================
body_types = ["Bipedal", "Quadruped", "Serpentine", "Floating", "Insectoid"]
traits = ["Highly Intelligent", "Aggressive", "Peaceful", "Curious", "Technological"]

def generate_species(home_planet):
    return {
        "name": alien_word(),
        "body": random.choice(body_types),
        "trait": random.choice(traits),
        "homeworld": home_planet
    }

# =========================
# SOLAR SYSTEM GENERATOR
# =========================
def generate_system():
    star_name = alien_word()
    num_planets = random.randint(3, 7)

    planets = [generate_planet() for _ in range(num_planets)]
    species = [generate_species(p["name"]) for p in random.sample(planets, k=2)]

    print("üåü STAR SYSTEM:", star_name)
    print("\nü™ê PLANETS:")
    for p in planets:
        print(f"- {p['name']} | {p['type']} | Moons: {p['moons']}")

    print("\nüëΩ INTELLIGENT SPECIES:")
    for s in species:
        print(f"- {s['name']} ({s['body']})")
        print(f"  Trait: {s['trait']}")
        print(f"  Homeworld: {s['homeworld']}\n")

# =========================
# RUN IT
# =========================
generate_system()
```

```python
import random

# =========================
# ADVANCED ALIEN LANGUAGE
# =========================

vowels = ["a", "e", "i", "o", "u", "aa", "ei", "ou"]
consonants = ["k", "z", "r", "x", "l", "t", "v", "sh", "m", "n", "q"]

def alien_word(min_parts=2, max_parts=4):
    word = ""
    parts = random.randint(min_parts, max_parts)

    for _ in range(parts):
        pattern = random.choice(["CV", "CVC", "VC"])
        if pattern == "CV":
            word += random.choice(consonants) + random.choice(vowels)
        elif pattern == "CVC":
            word += random.choice(consonants) + random.choice(vowels) + random.choice(consonants)
        else:
            word += random.choice(vowels) + random.choice(consonants)

        # small chance for apostrophe
        if random.random() < 0.2:
            word += "'"

    return word.capitalize().strip("'")

    # =========================
# ALIEN ALPHABET
# =========================

alien_alphabet = {
    "a": "‚àÜ", "e": "‚àë", "i": "‚àè", "o": "Œ©", "u": "¬µ",
    "aa": "‚àÜ‚àÜ", "ei": "‚àë‚àÜ", "ou": "Œ©¬µ",

    "k": "‚üä", "z": "‚ü°", "r": "‚ü¢", "x": "‚úï",
    "l": "‚åñ", "t": "‚ä¢", "v": "‚åó", "sh": "œû",
    "m": "‚üò", "n": "‚åá", "q": "‚üê",

    "'": "¬∑"
}

```

```python
def alien_translate(word):
    result = ""
    i = 0
    word = word.lower()

    while i < len(word):
        # check 2-letter sounds first (aa, ei, sh, etc.)
        if i + 1 < len(word) and word[i:i+2] in alien_alphabet:
            result += alien_alphabet[word[i:i+2]]
            i += 2
        elif word[i] in alien_alphabet:
            result += alien_alphabet[word[i]]
            i += 1
        else:
            i += 1  # skip unknown characters

    return result
```

```python
# =========================
# HUMAN ‚Üí ALIEN TRANSLATOR
# =========================

human_to_alien_map = {
    "a": "ka", "b": "zu", "c": "ra", "d": "xi", "e": "lo",
    "f": "ta", "g": "vek", "h": "shi", "i": "un", "j": "qa",
    "k": "ko", "l": "li", "m": "mor", "n": "na", "o": "ou",
    "p": "pi", "q": "qi", "r": "ra", "s": "sa", "t": "tar",
    "u": "uu", "v": "va", "w": "wo", "x": "xa", "y": "ya", "z": "zu",
    " ": " "
}

def human_to_alien(text):
    alien = ""
    for char in text.lower():
        alien += human_to_alien_map.get(char, "")
    return alien.capitalize()
```

```python
phrase = "Hello universe"
alien_phrase = human_to_alien(phrase)

print("Human:", phrase)
print("Alien (spoken):", alien_phrase)
print("Alien (written):", alien_translate(alien_phrase))
```

```python
ALIEN_ALPHABETS = [
    "‚àÜŒ©Œ®ŒõŒ¶Œ£Œû‚äó‚äï",
    "‚åò‚å¨‚å≠‚åÆ‚åØ‚å∞‚å±",
    "‚ú∂‚ú∑‚ú∏‚úπ‚ú∫‚úª‚úº",
    "‚üÅ‚üí‚üü‚ü°‚ü£‚ü§‚ü•"
]
```

```python
def alien_word():
    syllables = ["ka", "zu", "ra", "lo", "xi", "un", "tar", "vek", "shi"]
    return "".join(random.choice(syllables) for _ in range(random.randint(2, 4))).capitalize()

def alien_translate(word, alphabet):
    symbols = list(alphabet)
    mapping = {chr(97+i): symbols[i % len(symbols)] for i in range(26)}
    return "".join(mapping.get(c.lower(), c) for c in word)
```

```python
def generate_planet():
    return {
        "name": alien_word(),
        "type": random.choice(["Desert", "Ocean", "Ice", "Lava", "Forest"]),
        "moons": random.randint(0, 4)
    }
```

```python
def generate_species(homeworld):
    alphabet = random.choice(ALIEN_ALPHABETS)
    name = alien_word()
    return {
        "name": name,
        "script": alphabet,
        "trait": random.choice(["Telepathic", "Winged", "Aquatic", "Energy-based"]),
        "homeworld": homeworld
    }
```

```python
def evolve_language(word, planet_type):
    if planet_type == "Ice":
        return word.replace("a", "i")
    if planet_type == "Lava":
        return word.replace("l", "x")
    return word
```

```python
def generate_system():
    star = alien_word()
    planets = [generate_planet() for _ in range(random.randint(3, 6))]
    species = [generate_species(p["name"]) for p in random.sample(planets, 2)]

    print("üåü STAR SYSTEM")
    print(star)
    print()

    print("ü™ê PLANETS")
    for p in planets:
        print(f"- {p['name']} ({p['type']}, Moons: {p['moons']})")

    print("\nüëΩ SPECIES")
    for s in species:
        evolved = evolve_language(s["name"], next(p["type"] for p in planets if p["name"] == s["homeworld"]))
        print(f"\nSpecies: {s['name']}")
        print("Alien Script:", alien_translate(evolved, s["script"]))
        print("Trait:", s["trait"])
        print("Homeworld:", s["homeworld"])
```

```python
ERA_EVENTS = [
    "discovered faster-than-light travel",
    "nearly went extinct due to a stellar disaster",
    "unified under a single planetary council",
    "was enslaved by a machine intelligence",
    "abandoned their original homeworld",
    "ascended beyond physical form",
    "triggered a galaxy-wide conflict"
]

MYTHS = [
    "born from the light of a dying star",
    "crafted by ancient cosmic beings",
    "evolved in the shadow of a black hole",
    "emerged from the oceans during an endless storm",
    "were once energy before choosing flesh"
]

VALUES = [
    "knowledge above all",
    "survival through adaptation",
    "collective consciousness",
    "honor through exploration",
    "balance with the universe"
]
```

```python
def generate_backstory(species, planet_type):
    myth = random.choice(MYTHS)
    event1, event2 = random.sample(ERA_EVENTS, 2)
    value = random.choice(VALUES)

    story = f"""
We are the {species['name']}.
We originated on the {planet_type.lower()} world of {species['homeworld']}.

Our kind believes we were {myth}.
In our early age, we {event1}.
Much later, we {event2}.

These events shaped our belief in {value}.
Even now, our civilization carries the memory of those ages
etched into our language and symbols.
"""
    return story
```

```python
import random

# =========================
# PLANETS
# =========================
planets = [
    {"name": "Kora", "type": "Ocean World"},
    {"name": "Virex", "type": "Desert World"},
    {"name": "Thalos", "type": "Ice World"},
    {"name": "Nyra", "type": "Super-Earth"},
    {"name": "Zyph", "type": "Gas Giant Moon"}
]

# =========================
# SPECIES GENERATOR
# =========================
def generate_species():
    return {
        "name": alien_word(),
        "homeworld": random.choice(planets)["name"],
        "trait": random.choice([
            "Highly intelligent",
            "Hive-minded",
            "Nomadic",
            "Bioluminescent",
            "Cybernetic"
        ]),
        "script": "Glyphic"
    }

# Generate species list
species = [generate_species() for _ in range(3)]
```

```python
def alien_translate(word, alphabet=alien_alphabet):
    result = ""
    i = 0
    word = word.lower()

    while i < len(word):
        if i + 1 < len(word) and word[i:i+2] in alphabet:
            result += alphabet[word[i:i+2]]
            i += 2
        elif word[i] in alphabet:
            result += alphabet[word[i]]
            i += 1
        else:
            i += 1

    return result
```

```python
for s in species:
    planet_type = next(
        p["type"] for p in planets if p["name"] == s["homeworld"]
    )

    print("\nüëΩ SPECIES REPORT")
    print("Name:", s["name"])
    print("Alien Script:", alien_translate(s["name"]))
    print("Trait:", s["trait"])
    print("Homeworld:", s["homeworld"])

    print("\nüìú BACKGROUND STORY")
    print(generate_backstory(s, planet_type))
```

```python
import matplotlib.pyplot as plt
import random

def draw_alien_glyph(word):
    fig, ax = plt.subplots(figsize=(4, 4))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.axis("off")

    x, y = 1, 5

    for char in alien_translate(word):
        size = random.uniform(0.5, 1.2)

        if char in ["‚àÜ", "‚àë", "Œ©"]:
            ax.plot([x, x+size], [y, y+size], linewidth=2)
        elif char in ["‚üä", "‚ü¢", "‚åñ"]:
            circle = plt.Circle((x, y), size/2, fill=False)
            ax.add_patch(circle)
        else:
            ax.scatter(x, y, s=80)

        x += 1.2

    plt.show()
```

```python
def generate_timeline(species_name):
    year = 0
    timeline = []

    for _ in range(5):
        year += random.randint(200, 1500)
        event = random.choice(TIMELINE_EVENTS)
        timeline.append(f"Year {year}: The {species_name} {event}.")

    return timeline
```
